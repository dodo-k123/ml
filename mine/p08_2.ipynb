### P08_2, P08_3 - Logistic Regression and Multi Logistic Regression

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
```

```python
iris_dataset = datasets.load_iris()
iris_features = iris_dataset.data
iris_target = iris_dataset.target
```

```python
scaler = StandardScaler()
features_std = scaler.fit_transform(iris_features)
logistic_regression = LogisticRegression(random_state=0)
model = logistic_regression.fit(features_std, iris_target)
```

```python
new_observation=[[.5,.5,.5,.5]]
print(model.predict(new_observation))
print(model.predict_proba(new_observation))
```

## Multi Logistic Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
```

```python
iris_dataset = datasets.load_iris()
iris_features = iris_dataset.data
iris_target = iris_dataset.target

scaler = StandardScaler()
features_std = scaler.fit_transform(iris_features)
logistic_regression = LogisticRegression(random_state = 0, multi_class = "ovr")
model = logistic_regression.fit(features_std, iris_target)
```

```python
new_observation=[[.8,.5,.8,.2]]
print(model.predict(new_observation))
print(model.predict_proba(new_observation))
```
